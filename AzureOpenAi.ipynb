```python
import requests
from bs4 import BeautifulSoup
```

```python
from posixpath import sep
def extract_text_from_url(url):
  response = requests.get(url)

  if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    (script_or_style.decompose() for script_or_style in soup(['script', 'style']))

    text = soup.get_text(separator=' ', strip=True)
    lines = (line.strip() for line in text.splitlines())
    chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
    text = '\n'.join(chunk for chunk in chunks if chunk)
    return text
  else:
    print(f'Failed to fetch URL: {url}')
```

```python
extract_text_from_url("https://www.nytimes.com/international/")
```

```python
from langchain_openai.chat_models.azure import AzureChatOpenAI

client = AzureChatOpenAI(
    azure_endpoint="",
    api_key="",
    deployment_name="gpt-4o-mini",
    api_version="2024-02-15-preview",
    max_retries=0
)

def translate_article(text, language):
  messages = [
      ("system",  "Você atua como um tradutor de textos"),
      ("user",  f"Traduza o seguinte {text} para o idioma {language} e responda em markdown")
  ]

  response = client.invoke(messages)
  print(response.content)
  return response.content

translate_article("Despite Covid ‘Amnesia,’ the Pandemic Simmers Beneath the 2024 Race", "pt-br")
```

```python
url = "https://dev.to/devteam/top-7-featured-dev-posts-of-the-week-36o4"
text = extract_text_from_url(url)
#print(text)
article = translate_article(text, "pt-br")
print(article)
```
